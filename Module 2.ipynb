## Task 1 - Plotting
* Plot the values of occupancy percentage versus number of weeks since release, for all movies in the dataset. 
* Plot the number of other releases in the week versus occupancy percentage and weeks until the end of lifetime, for all movies.
* Plot the number of shows in the week versus occupancy percentage and weeks until the end of lifetime, for all movies.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("Data/CinemaData.csv")
```

```python
print(df)
```

```python
print(df.keys())
df.iloc[,7]
```

```python
plt.scatter(x=df['WeeksSinceRelease'],y=df['OccPer'])
plt.ylabel('OccupancyPercentage')
plt.xlabel('WeeksSinceRelease')
```

```python
plt.scatter(x=df['OccPer'],y=df['OtherReleasesInWeek'],color='red')
plt.xlabel('OccupancyPercentage')
plt.ylabel('OtherReleasesInWeek')
plt.show()
```

```python
plt.scatter(x=df['Lifetime']-df['WeeksSinceRelease'],y=df['OtherReleasesInWeek'],color='green')
plt.ylabel('OtherReleasesInWeek')
plt.xlabel('WeeksUntilEndOfLifetime')
plt.show()
```

```python
plt.scatter(x=df['OccPer'],y=df['ShowsInWeek'],color='black')
plt.ylabel('ShowsInWeek')
plt.xlabel('OccupancyPercentage')
plt.show()
```

```python
x=[df.iloc[i,4]-df.iloc[i,11] for i in range(len(df))]

plt.scatter(x,df['ShowsInWeek'],color='violet')
plt.ylabel('ShowsInWeek')
plt.xlabel('WeeksUntilEndOfLifetime')
plt.show()
```
